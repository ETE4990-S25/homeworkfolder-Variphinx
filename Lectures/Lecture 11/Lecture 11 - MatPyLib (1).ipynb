## Sub Plots ( the flollowing are examples provided by ChatGPT)

```python
#stolen Example and modified for class
import matplotlib.pyplot as plt
import numpy as np # numpy introduces some amazing tools to be talked about next

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with 2 subplots
fig, axs = plt.subplots(2,tight_layout=True)

# First subplot
axs[0].plot(x, y1, 'r')
axs[0].set_title('Sine Wave')
axs[0].set(xlabel= "distance(m)", ylabel="amplitude")
axs[0].grid()

# Second subplot
axs[1].plot(x, y2, 'b')
axs[1].set_title('Cosine Wave')
axs[1].set(xlabel= "distance(m)", ylabel="amplitude")

# Add labels and titles
fig.suptitle('Advanced Subplots')
#plt.xlabel('X-axis') #note this will overwrite the previous xlabel
#plt.ylabel('Y-axis')

# Display the plot

plt.show()


```

## 3d Plots 
We will not deal with 3d plots but the below is nice to know


```python
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Add labels and title
ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Display the plot
plt.show()
```

```python
# You will need to install mplcursors for this example and run it as a py file
import matplotlib.pyplot as plt
import mplcursors
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y)

# Add interactive cursors
mplcursors.cursor(ax, hover=True)

# Add title and labels
plt.title('Interactive Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)
y4 = np.sin(x)**2

# Create a grid of 2x2 subplots with shared axes
fig, axs = plt.subplots(2, 2, sharex='col', sharey='row')

# First subplot
axs[0, 0].plot(x, y1, 'r')
axs[0, 0].set_title('Sine')

# Second subplot
axs[0, 1].plot(x, y2, 'g')
axs[0, 1].set_title('Cosine')

# Third subplot
axs[1, 0].plot(x, y3, 'b')
axs[1, 0].set_title('Sine * Cosine')

# Fourth subplot
axs[1, 1].plot(x, y4, 'k')
axs[1, 1].set_title('Sine^2')

# Add a main title
fig.suptitle('Advanced Subplots with Shared Axes')

# Adjust layout to prevent overlap
```

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(x)

fig, ax1 = plt.subplots()

# Plotting the first y-axis
ax1.plot(x, y1, 'g-')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Sine', color='g')

# Adding a secondary y-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, 'b-')
ax2.set_ylabel('Exponential', color='b')

# Add title
plt.title('Plot with Secondary Y-axis')

# Display the plot
plt.show()
```

```python
# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Create the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

# Add title
plt.title('Pie Chart with Annotations')

# Display the plot
plt.show()
```

# Live Plots

```python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
ax.axis([0,2*np.pi,-1,1])
l, = ax.plot([],[])

def animate(i):
    l.set_data(t[:i], x[:i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))

from IPython.display import HTML
HTML(ani.to_jshtml())
```
